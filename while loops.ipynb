# Fibonacci series

```python
number =  int(input("Enter  the  number :-"))
a, b = 0, 1
counter = 0
while counter<number:
  print(a)
  c=a+b
  a= b
  b=c
  counter = counter+1
```

```python
word   =   input("enter  the  string")
reverse =  ""
length =  len(word)
for i in range(length-1,-1,-1):
  reverse = reverse+word[i]
print(reverse)
```

```python
word   =   input("enter  the  string")
reverse =  ""
length =  len(word)
while length >0:
  reverse = reverse+word[length-1]
  length = length-1
print(reverse)
```
