*Diffrence*  between  shallow  copy  vs  deep  copy

```python
#-Shallow  copay  it  indeicate  to  different object
a =  [1,2,3,44,5]
b =  a.copy()
print(a)
print(b)
print(id(a))
print(id(b))
```

```python
# deep copy  it indeicate  the  same  object
a =  [1,2,3,44,5]
b =  a
print(a)
print(b)
print(id(a))
print(id(b))
```
