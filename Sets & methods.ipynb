```python
s =  {1,2,3,45,5}
s.add(4)
s
```

**set.update()**


*   update  method  update  the  set  with  another  set  or  any iterable



```python
s = {1,2,3,4}
s1 = {5,6,7}
s.update([("a","aa"),"b",5])
print(s)
```

**set.discard()**

*   discard() method in  the  set remove  the  specific  element  from  the  set  if  element  does  not  exist  in  the  set  nothing  will  happens



```python
s = {1,2,3,4,5}
s.discard(3)
print(s)
```

**Set.clear()**


*   clear  method  remove  all  elemnt  from  the  set



```python
s = {1,2,3,4,5}
s.clear()
print(s)#  remove  all  items  in  the set
```

```python
s = {1,2,3,4}
del  s            # deleted the  variable
print(s)
```

** set.copy()**


*   c poy  method  return  a  shallow  copy  of  set
*   cahnge  to  the  copies  set  does't affect  the  orignal  set



```python
a = {1,2,3}
b = a.copy()
b.add(4)
print(a)
print(b)
```

**Set.union**

*   this  method  return  a  new  set  with  elemnt  from  both  sets




```python
s1 = {1,2,3,4}
s2 = {4,5,6,7}
s3 = s1.union(s2)
print(s3)
```

**se.Intersection**


*   insection  method  return  a  new  set  with  selemt  that  are  commen  to  all  set


```python
s1 = {1,2,3,4}
s2 = {4,3,4,7}
s3= s1.intersection(s2)
print(s3)
```

**Set.disjoint()**

*   disjoint  method  return  true  if two  sets  have  no  intersection  otherwise  it  return  false



```python
s1 = {1,2,4,5}
s2 = {5,6,7,8}
s3 = {9,10,11}
print(s1.isdisjoint(s2))

```

**set.diffrence()**


*   diffrence  method  return  a  new  set  with  element  in  the  set  that  are  in  the  other  specific  sets



```python
s1 = {1,2,3,4}
s2 = {2,3,4,5}
print(s1-s2)
print(s2-s1)
```

**issubset()**

*   methos  return  true  if  all  elemnt  of  first  set  are  available  in  the  second  set ,  otherwise  it return  false


```python
a =  {3,4,5,6}
b = {5,6,7,8}

ans = a.issubset(b)
print(ans)

ans = b.issubset(a)
print(ans)
```
