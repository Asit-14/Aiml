# Dictinaries

*   in python  dic  are  difiend  within  {} with  each  item  being key:values
*   key  can  be of  fundamental data  types  -  generally  the key  are  int  or  str
*   values  can  be  of  any  data  types


empty  dic

````python
dic = {}
print(dic)
````

dic with  1  item

````python
d = {'a':1}
print(d)
````

dic  with  two  item

````python
d = {'a':1 , 'b':2}
print(d)
````
